Question 1
Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def merge_linked_lists(self,head1, head2):
    l1 = head1
    l2 = head2
    mergedHead = None
    mergedTail = None

    while l1 is not None and l2 is not None:
        if l1.data >= l2.data:
            new_node = Node(l1.data)
            l1 = l1.next
        else:
            new_node = Node(l2.data)
            l2 = l2.next

        if mergedHead is None:
            mergedHead = new_node
            mergedTail = new_node
        else:
            mergedTail.next = new_node
            mergedTail = mergedTail.next

        return mergedHead
```

Question 2

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def remove_duplicates(head):
    current = head

    while current is not None and current.next is not None:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head
```

Given a linked list of size **N**. The task is to reverse every **k** nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple of *k* then left-out nodes, in the end, should be considered as a group and must be reversed.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_k_nodes(head, k):
    current = head
    prev = None
    next = None
    count = 0

    # Reverse first k nodes
    while current is not None and count < k:
        next = current.next
        current.next = prev
        prev = current
        current = next
        count += 1

    # If there are more nodes, recursively reverse the remaining
    if next is not None:
        head.next = reverse_k_nodes(next, k)

    return prev
```

```python
head = Node(1)
node1 = Node(2)
node2 = Node(2)
node3 = Node(4)
node4 = Node(5)
node5 = Node(6)
node6 = Node(7)
node7 = Node(8)
node8 = Node(9)

head.next = node1
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7
node7.next = node8

# Reverse nodes in groups of k

k = 4
head = reverse_k_nodes(head, k)

# Print the modified linked list

ptr = head
while ptr is not None:
    print(ptr.data, end=" -> ")
    ptr = ptr.next
```

Question 4

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_alternate_k_nodes(head, k):
    current = head
    prev = None
    next = None
    count = 0

    # Reverse alternate k nodes
    while current is not None and count < k:
        next = current.next
        current.next = prev
        prev = current
        current = next
        count += 1

    # Connect the reversed portion to the appropriate nodes
    if head is not None:
        head.next = current

    # Skip k-1 nodes in the remaining list
    count = 0
    while count < k-1 and current is not None:
        current = current.next
        count += 1

    # Recursively reverse the next alternate k nodes
    if current is not None:
        current.next = reverse_alternate_k_nodes(current.next, k)

    return prev
 # The time complexity of this algorithm is O(n)
```

```python
# Create the linked list

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)
node5 = Node(6)
node6 = Node(7)
node7 = Node(8)
node8 = Node(9)

head.next = node1
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7
node7.next = node8

# Reverse every alternate k nodes efficiently

k = 3
head = reverse_alternate_k_nodes(head, k)

# Print the modified linked list

ptr = head
while ptr is not None:
    print(ptr.data, end=" -> ")
    ptr = ptr.next
```

</aside>

```python
# Node class definition

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to delete the last occurrence of a key

def delete_last_occurrence(head, key):
    prev = None
    last = None
    current = head

    # Find the last occurrence of the key
    while current is not None:
        if current.data == key:
            prev = last
            last = current
        current = current.next

    # If the key was not found
    if last is None:
        return head

    # Delete the last occurrence of the key
    if last == head:
        head = last.next
    else:
        prev.next = last.next

    return head

# Create the linked list

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(3)
node5 = Node(5)
node6 = Node(6)

head.next = node1
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6

# Delete the last occurrence of the key

key = 3
head = delete_last_occurrence(head, key)

# Print the modified linked list

ptr = head
while ptr is not None:
    print(ptr.data, end=" -> ")
    ptr = ptr.next
```

Question 6

Given two sorted linked lists consisting of **N** and **M** nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def merge_sorted_lists(head1, head2):
    dummy = Node(0)  # Dummy node
    current = dummy
    while head1 is not None and head2 is not None:
        if head1.data <= head2.data:
            current.next = head1
            head1 = head1.next
        else:
            current.next = head2
            head2 = head2.next
        current = current.next

    # Connect remaining nodes from list1 or list2
    if head1 is not None:
        current.next = head1
    if head2 is not None:
        current.next = head2

    return dummy.next

```

```python
# Create the first sorted linked list
head1 = Node(1)
head1.next = Node(3)
head1.next.next = Node(5)
head1.next.next.next = Node(7)

# Create the second sorted linked list
head2 = Node(2)
head2.next = Node(4)
head2.next.next = Node(6)
head2.next.next.next = Node(8)

# Merge the two sorted lists
merged_head = merge_sorted_lists(head1, head2)

# Print the merged list
ptr = merged_head
while ptr is not None:
    print(ptr.data, end=" -> ")
    ptr = ptr.next
```

Question 7

Given a **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverse_doubly_linked_list(head):
    current_node = head
    previous_node = None

    while current_node is not None:
        next_node = current_node.next

        # Swap previous and next pointers of the current node
        current_node.next = previous_node
        current_node.prev = next_node

        previous_node = current_node
        current_node = next_node

    # Set the last node visited as the new head
    head = previous_node

    return head
```

Question 8

Given a doubly linked list and a position. The task is to delete a node from given position in a doubly linked list.

</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def delete_node_at_position(head, position):
    if head is None:
        return head

    if position == 0:
        new_head = head.next
        if new_head:
            new_head.prev = None
        return new_head

    current_node = head
    count = 0

    while current_node is not None and count < position:
        current_node = current_node.next
        count += 1

    if current_node is None:
        return head

    previous_node = current_node.prev
    next_node = current_node.next

    if previous_node:
        previous_node.next = next_node

    if next_node:
        next_node.prev = previous_node

    return head
```
